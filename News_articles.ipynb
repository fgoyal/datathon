```python
from sklearn.model_selection import train_test_split
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sacremoses import MosesTokenizer
```

```python
nltk.download('stopwords')
vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'), lowercase=True, min_df=2, max_df=0.5, max_features=5000)
```

```python
mt = MosesTokenizer()

tokenized = [mt.tokenize(x) for x in data['content']]
tokenized = np.array(tokenized)

data['tokenized'] = tokenized
```

```python
data['tokenized']
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
nltk.download('vader_lexicon')

sid = SentimentIntensityAnalyzer()

def get_polarity_scores_neg(text):
    return sid.polarity_scores(text)['neg']
def get_polarity_scores_neu(text):
    return sid.polarity_scores(text)['neu']
def get_polarity_scores_pos(text):
    return sid.polarity_scores(text)['pos']
def get_polarity_scores_compound(text):
    return sid.polarity_scores(text)['compound']

data['neg'] = data['content'].apply(get_polarity_scores_neg)
data['neu'] = data['content'].apply(get_polarity_scores_neu)
data['pos'] = data['content'].apply(get_polarity_scores_pos)
data['compound'] = data['content'].apply(get_polarity_scores_compound)
```
